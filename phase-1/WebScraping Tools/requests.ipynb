***REQUEST***

Send HTTP requests to websites and get their data.

***1. GET :*** Retrieve data from a server.

```python
import requests
response = requests.get('https://github.com/chennakesava-pinnamasetty')

if response.status_code == 200:   # 200 means successfull
    print("Page loaded successfully")
else:
    print("Error occurred")
    
```

```python
print(response.content)
```

***2. POST :***  send data to the server, typically for creating or submitting resources.

```python
import requests

r = requests.post('https://httpbin.org//post', data ={'key':'value'})

# success code - 200
print(r)

# print content of request
print(r.json())
```

```python
import requests

r = requests.get("https://api.github.com/users/chennakesava-pinnamasetty")

print(r.status_code)
print(r.json())
```

***3. PUT :*** PUT is used to update or replace an entire resource on the server.

```python
import requests

res = requests.put('https://httpbin.org/put', data={'key': 'value'})

print("Status Code:", res.status_code)

print("Response Body:", res.content.decode())
```

***PUT Ex:***

```python
import requests

res = requests.put(
    "https://httpbin.org/put",
    json={"name": "Chenna", "age": 23}
)

print("Status Code:", res.status_code)
print("Response Body:", res.json())
```

***4. PATCH :*** We can change Specific fields we want

```python
import requests

res = requests.patch(
    "https://httpbin.org/patch",
    json={"data": 23}
)

print(res.status_code)
print(res.json())
```

***5. DELETE :*** Delete specific one in the server

```python
import requests

r = requests.delete('https://httpbin.org//delete', data ={'key':'value'})
print(r)

print(r.json())
```

***6. HEAD :*** HEAD is an HTTP method used to retrieve only the headers of a response  not the body.

GET  Returns headers + body

HEAD  Returns only headers

```python
import requests

r = requests.head('https://httpbin.org/', data ={'key':'value'})

print(r)

print(r.headers)

print(r.content)
```
