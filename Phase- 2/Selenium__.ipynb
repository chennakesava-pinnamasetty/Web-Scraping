```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

driver = webdriver.Chrome()
driver.maximize_window()

wait = WebDriverWait(driver, 10)

driver.get("https://www.indiabix.com/")


wait.until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, "card-style-list"))
)

sections = driver.find_elements(By.CLASS_NAME, "card-style-list")

data_dict = {}

for section in sections:
    lines = section.text.split("\n")
    
    if lines:
        heading = lines[0].strip()
        items = [item.strip() for item in lines[1:] if item.strip()]
        
        data_dict[heading] = items


df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in data_dict.items()]))

# Save to CSV
df.to_csv("IndiaBix_Homepage.csv", index=False)

print("CSV file created successfully ✅")

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

driver = webdriver.Chrome()
driver.maximize_window()

wait = WebDriverWait(driver, 10)

driver.get("https://www.indiabix.com/")

# ==========================
# 1️⃣ HOMEPAGE DATA
# ==========================

wait.until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, "card-style-list"))
)

sections = driver.find_elements(By.CLASS_NAME, "card-style-list")

data_dict = {}

for section in sections:
    lines = section.text.split("\n")
    
    if lines:
        heading = lines[0].strip()
        items = [item.strip() for item in lines[1:] if item.strip()]
        data_dict[heading] = items


# ==========================
# 2️⃣ CLICK VERBAL ABILITY
# ==========================

title = wait.until(
    EC.element_to_be_clickable((By.LINK_TEXT, "Verbal Ability"))
)
title.click()

verbal_topics = wait.until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".need-ul-filter a"))
)

verbal_list = [t.text.strip() for t in verbal_topics if t.text.strip()]

# Add Verbal Ability inner topics as new column
data_dict["Verbal Ability (Inner Topics)"] = verbal_list


# ==========================
# 3️⃣ CONVERT TO DATAFRAME
# ==========================

df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in data_dict.items()]))

# Save to Excel
df.to_excel("IndiaBix_Complete_Data.xlsx", index=False)

print("Excel file created successfully ✅")

driver.quit()
```
