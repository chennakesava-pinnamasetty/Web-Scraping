```python


url = "https://edugo-api.aspiria.ai/aspiria/course-finder/"

headers = {
    "Accept": "application/json, text/plain, */*",
    "Origin": "https://aspiria.ai",
    "Referer": "https://aspiria.ai/",
    "User-Agent": "Mozilla/5.0"
}


params = {
    "page": 1,
    "page_size": 16
}

response = requests.get(url, headers=headers, params=params)
data = response.json()

total_pages = data["data"]["pagination"]["total_pages"]

print("Total Pages:", total_pages)


for page in range(1, total_pages + 1):

    print(f"\nFetching Page {page}...\n")

    params["page"] = page

    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    courses = data["data"]["university_courses"]

    for item in courses:

       
        print("Course Name :", item['title'])
        print("Country     :", item['country'])
        print("City        :", item['city'])
        print("TuitionFees :", item['tuitionFees'])
        print("Duration    :", item['duration'])
        print("University  :", item['university'])
        print("-" * 40)

    time.sleep(0.5)  

print("\n All pages data printed successfully.")
```
